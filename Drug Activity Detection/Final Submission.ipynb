```python
import pandas as pd
import scipy
from scipy.sparse import csr_matrix
import numpy as np
from sklearn import svm
```

```python
#Read the train data and save it in dataframe after splitting by new lines
with open('train.dat','r') as f:
    df = pd.DataFrame(l.split("\t") for l in f) 
df.columns = ["result","features"]  

dff = pd.DataFrame(df["features"].str.split('\n'))
drr = pd.DataFrame(df["result"].str.split('\n'))
dt = df['features']
dr = df['result']
```

```python
arrs = np.empty(800, dtype=np.object)
count=0
for f in dff['features']:
        for f1 in f:
            arr =[]
            t = f1.rstrip(' ')
            k = t.rstrip(',')
            r = k.split(' ')
            for i in r:
                if(len(i)>0 and int(i)<=70000):
                    arr.append(i)
            if(len(arr)>1):
                arr = [int(i) for i in arr]
                arr1 = np.asarray(arr)
            arrs[count] = arr1 
        count += 1
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=10)
drm = dr.as_matrix()
drm = [int(i) for i in drm]
x,y = arrs,drm
xlist = [[]*800 for x in range(800)]
y1 = np.asarray(y)
for i in range(800):
    lst = [0] * 70000
    for a in x[i]:
        lst[a-1] = 1     
    x1 = np.asarray(lst)
    xlist[i] = x1
    i += 1 
modX = np.array(xlist)
print("Got data and now fitting into model")
clf.fit(modX,y1)
print("Model Trained!")
```

```python
#Test data processing and finally classifying them using previously build model
with open('test.dat','r') as f:
    dft = pd.DataFrame(l.split("\t") for l in f) 
dft.columns = ["features"]  
dfft = pd.DataFrame(dft["features"].str.split('\n'))
output = open("prediction_random_clf.dat", 'w') 
for f in dfft["features"]:
        for f1 in f:
            arr =[]
            t = f1.rstrip(' ')
            k = t.rstrip(',')
            k = t.rstrip('\n')
            r = k.split(' ')
            for i in r:
                if(len(i)>0 and int(i)<=70000):
                    arr.append(i)
            if(len(arr)>1):
                arr = [int(i) for i in arr]
                arr1 = np.asarray(arr)  
        lst = [0] * 70000
        for a in arr1:
            lst[a-1] = 1
        x1 = np.asarray(lst)
        if(len(x1)>1):
            res = clf.predict(x1)
            if(res[0] == 0):
                output.write("0\n")
            elif(res[0] == 1):
                output.write("1\n")
output.close()
print("Prediction Done, Check Output file!")
```
